### First chord in open position

```python
path = "./csv/chordDataO00.csv"

dfO = pd.read_csv(path, index_col = 0);

dfO.columns = pd.MultiIndex.from_product([["Next Chord"], dfO.columns])
dfO.index = pd.MultiIndex.from_product([["First Chord"], dfO.index])

dfO
```

```python
a = (dfO == 0).sum().sum()
b = (dfO > 10).sum().sum()

a, b, dfO.size
```

```python
dfO.max(axis=1)
```

```python
zero_columns_per_row = dfO.apply(
    lambda row: [col[1] for col in row.index[row == 0].tolist()],
    axis = 1
)

zero_columns_per_row
```
